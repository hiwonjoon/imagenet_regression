```python
class tree :
    def __init__(self,wnid) :
        self.wnid = wnid
        self.parents = []
        self.children = []
    def add_parent(self,parent) :
        self.parents.append(parent)
    def add_child(self,child) :
        self.children.append(child)
    def __str__(self):
        return self.wnid
nodes_dict = {}
```

```python
for line in open('../wordnet.is_a.txt').readlines() :
    parent_id, child_id = line.strip().split(' ')
    if parent_id in nodes_dict :
        parent = nodes_dict[parent_id]
    else :
        parent = tree(parent_id)
        nodes_dict[parent_id] = parent
    
    if child_id in nodes_dict :
        child = nodes_dict[child_id]
    else :
        child = tree(child_id)
        nodes_dict[child_id] = child
    
    child.add_parent(parent)
    parent.add_child(child)
```

```python
from sets import Set
def get_parents(node,parents) :
    for parent in node.parents :
        if parent not in parents :
            get_parents(parent,parents)
    parents.add(node)
    
synset_list_2012 = open('../synsets.txt').readlines()
synset_list_2012 = [wnid.strip() for wnid in synset_list_2012]

import itertools
import numpy as np
distance_matrix = np.zeros((1000,1000),np.uint32)

for i in range(1000) :
    for j in range(i,1000) :
        p1 = Set(); p2 = Set();
        get_parents(nodes_dict[synset_list_2012[i]],p1)
        get_parents(nodes_dict[synset_list_2012[j]],p2)

        distance_matrix[i,j] = len( p1.symmetric_difference(p2) )
        distance_matrix[j,i] = distance_matrix[i,j]
```

```python
synset_words = open('../synset_words.txt').readlines()
synset_words = [name.strip() for name in synset_words]

check = np.argsort(distance_matrix.reshape(1000*1000))[::-1]
for (k,ith) in enumerate(check) :
    (i,j) = np.unravel_index(ith,(1000,1000))
    if( i == j ) :
        continue
    print synset_words[i], synset_words[j], distance_matrix[i,j]
    if( k > 1500 ) :
        break
# Maxium Distance is 36
```
